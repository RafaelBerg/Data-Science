É possivel carregar os DataFrames diretamente do site utilizando o read_csv, mas ajustando a funcionalidade do comando para que isso seja possivel

Carregando os arquivos em DataFrames

```python
df_2018 = pd.read_csv("https://arquivos.prf.gov.br/arquivos/index.php/s/MaC6cieXSFACNWT/download",compression="zip",encoding="latin1",sep=";",low_memory=False)
df_2019 = pd.read_csv("https://arquivos.prf.gov.br/arquivos/index.php/s/kRBUylqz6DyQznN/download",compression="zip",encoding="latin1",sep=";",low_memory=False)
df_2020 = pd.read_csv("https://arquivos.prf.gov.br/arquivos/index.php/s/jdDLrQIf33xXSCe/download",compression="zip",encoding="latin1",sep=";",low_memory=False)

# usando concacat, pois se trata de arquivos com as mesmas colunas e parâmetros
# adcionando então embaixo do proximo DataFrame
df_juntos = pd.concat([df_2018,df_2019,df_2020])
```

2) (a) quantos acidentes ocorreram em cada ano?

<h1> De Ilesos por ano</h1>

Ano de 2018:

```python
meses_2018 = df_2018
meses_2018["data_inversa"] = pd.to_datetime(meses_2018["data_inversa"])

ilesos_2018 = pd.DataFrame(meses_2018.groupby(meses_2018["data_inversa"].dt.month)["ilesos"].sum())

pessoas_2018 = pd.DataFrame(meses_2018.groupby(meses_2018["data_inversa"].dt.month)["pessoas"].sum())

proporção_ilesos_2018 = pd.concat([ilesos_2018, pessoas_2018], axis=1)
proporção_ilesos_2018["proporção (ilesos/total)"] = proporção_ilesos_2018["ilesos"] / proporção_ilesos_2018["pessoas"]
proporção_ilesos_2018
```

(c) mostre, reorganizando o conjunto de dados, se a pandemia conseguiu diminuir ou não
a incidência de acidentes nas rodovias federais.

```python
df_18_19 = pd.concat([df_2018,df_2019], ignore_index=True)

df_2020_1 = df_2020
df_2020_1["mes"] = df_2020_1["data_inversa"].dt.month
df_2020_mes = (df_2020_1["mes"]>=3).sum()

df_2020_2 = df_2020
df_2020_2["mes"] = df_2020_2["data_inversa"].dt.month
df_2020_meses = (df_2020_2["mes"]<=2).sum()

media_before_pand = (len(df_18_19) + df_2020_meses) / 26
media_before_pand = round(media_before_pand,1)
media_after_pand = (df_2020_mes / 10)

df_media_acidentes = pd.DataFrame(data=[[media_before_pand,media_after_pand]], columns=[["Sem Pandemia", "Com pandemia"]])
df_media_acidentes
```

havendo uma diminuição de acidentes durante a pandemia
